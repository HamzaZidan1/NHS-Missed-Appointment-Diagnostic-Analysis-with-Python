### To review the hashtags, these steps need to be taken:

#### 1. Create a new DataFrame with only text values

```python
tweets_text = tweets['tweet_full_text']

print(tweets_text.head())
```

#### 2. Create an empty list for hashtags

```python
tags = []
```

#### 3. Extract hashtags from the tweet_full_text column and append them to the tags list

```python
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            tags.append(z.lower())
```

#### 4. Create a Pandas Series to count the values in the list

```python
tags_series = pd.Series(tags)


print(tags_series.value_counts().head(30))
```

### To identify the top trending hashtags with a visualisation, these steps need to be taken:

#### 1. Convert the new Series into a DataFrame

```python
data = pd.DataFrame({'word': tags_series.value_counts().index, 'count': tags_series.value_counts().values}).reset_index(drop=True)
data = data.rename(columns={'word': 'Hashtag', 'count': 'Count'})
data['Count'] = data['Count'].astype(int)

print(data.head())
```

#### 2. Create seaborn barplot with count > 10

```python
sns.barplot(x='Hashtag', y='Count', data=data[data['Count']>10])
plt.title('Top Trending Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

#### 4. Remove overrepresented hashtags

```python
# list the hashtags to remove
overrepresented = ['#healthcare', '#health', '#medicine'] 
data = data[~data['Hashtag'].isin(overrepresented)]
print(data.head(10))
```

#### 4. Create seaborn barplot of remaining most frequently used hashtags

```python
sns.barplot(x='Hashtag', y='Count', data=data.head(10))
plt.title('Top Trending Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
